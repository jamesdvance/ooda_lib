```python
from aiohttp import ClientSession
import asyncio
from blinkpy.blinkpy import Blink
from blinkpy.auth import Auth
from blinkpy.helpers.util import json_load
file_location = "/home/james/.auth/blink_auth.json"
```

```python
async def start():
    blink = Blink()
    auth = Auth(await json_load(file_location))
    blink.auth = auth
    await blink.start()
    return blink


def get_cameras(blink:Blink):
    names, cameras = [], []
    for name, camera in blink.cameras.items():
        names.append(name)
        cameras.append(camera)

    return names, cameras
```

```python
    cap = cv2.VideoCapture(RTSP_URL)

    if not cap.isOpened():
        print("Error: Could not open RTSP stream.")
    else:
        try:
            while True:
                ret, frame = cap.read()
                if not ret:
                    print("Error: Failed to read frame from stream.")
                    break

                # Convert the frame from BGR to RGB for Matplotlib display
                frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

                # Display the frame
                plt.imshow(frame_rgb)
                plt.axis('off') # Hide axes
                display(plt.gcf()) # Display the current figure
                plt.close() # Close the figure to avoid memory issues

                # Add a small delay to control frame rate and allow for interruption
                # You might need to adjust this depending on your stream and desired frame rate
                # For a more interactive experience, consider using ipywidgets for control
                # For continuous streaming, a more advanced approach might be needed
                # to avoid overwhelming the notebook output.
                # This simple loop will generate many outputs.
                # To stop, interrupt the kernel.
                # time.sleep(0.01) # Optional: Add a small delay
        except KeyboardInterrupt:
            print("Stream stopped by user.")
        finally:
            cap.release()
            cv2.destroyAllWindows()
```
